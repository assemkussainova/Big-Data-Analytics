```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import itertools

import networkx as nx
from networkx.algorithms.community.centrality import girvan_newman
```

```python
B = pd.read_csv('bda-graphtask-adj.csv', header=None)

G = nx.from_numpy_matrix(B.values)

plt.imshow(nx.adjacency_matrix(G).todense())
```

```python
# recover the structure of this network using clustering
comp = girvan_newman(G, most_valuable_edge = None)

k = 4
count = 0
clusters = []

for communities in itertools.islice(comp, k):
    if count == 3:
        for c in communities:
            nodelist = list(c)
            clusters.append(nodelist)
            print(nodelist)
    count += 1

pos = nx.spring_layout(G)
```

```python
colors = ['red', 'green', 'blue', 'yellow', 'black']

for i in range(5):
    nx.draw_networkx_nodes(G, pos, nodelist = clusters[i], node_color = colors[i], alpha = 0.8)

nx.draw_networkx_edges(G, pos)
plt.show()
```
