```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
for i in range(1, 9):
    # initialize chess board
    n = 2 ** i 
    board = np.zeros((8, 8)) 
    board[0, 0] = 1
    
    # run 1000 iterations to check different cases of king moves
    for j in range(1000):     
        start = [7, 0]
        steps = 0
        
        while steps < n:
            X = random.randint(-1, 1)
            Y = random.randint(-1, 1)     
            if ([X, Y] != [0, 0]) and (start[0] + X) < 8 and (start[0] + X) >= 0 and (start[1] + Y) < 8 and (start[1] + Y) >= 0:
                start[0] += X
                start[1] += Y
                steps += 1        
        board[start[0], start[1]] += 1
    
    # calculate probability for each cell
    total = board.sum()
    for height in range(8):
        for width in range(8):
            board[height, width] /= total
    
    # plot chess board with probability values
    plt.figure(i)
    plt.title('Image ' + str(i) + '. ' + 'Moves: ' + str(n))
    plt.imshow(board, cmap = "Greys")
    plt.colorbar()
```
